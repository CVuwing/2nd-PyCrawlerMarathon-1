### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('em')[-1].text, d.find_all('a')[-1].text)
```

### ③ 哪一個類別的文章最多

```python
var = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find_all('em')[-1].text not in var:
        var[d.find_all('em')[-1].text] = 1
        #print(var)
        #print(var[d.find_all('em')[-1].text])
        #print(type(var[d.find_all('em')[-1].text]))
    else:
        var[d.find_all('em')[-1].text] +=1
    #print(d.find(class_="date").text, d.find_all('em')[-1].text, d.find_all('a')[-1].text)
#print(max(var))
print('文章最多的類別：', max(var), '，有', str(var[max(var)]), '篇文章')
```
