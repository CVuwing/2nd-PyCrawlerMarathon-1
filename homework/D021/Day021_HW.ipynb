## 1. 取出今天所有的新聞

```python
# 打開瀏覽器
from selenium import webdriver
from bs4 import BeautifulSoup
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

# 每個兩秒鐘自動往下滑
import time
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")

# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
today = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()).split(' ')[0].split('-')[2] #抓今日
for d in soup.find(class_="part_list_2").find_all('h3'):
    dday = d.find(class_="date").text.split(' ')[0].split('/')[2]
    if dday == today:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    else:
        break

# 關閉瀏覽器
browser.quit();
```

## 2. 取出現在時間兩小時內的新聞

```python
# 打開瀏覽器
from selenium import webdriver
from bs4 import BeautifulSoup
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

# 每個兩秒鐘自動往下滑
import time
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")

# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
tomin = time.strftime("%M", time.localtime())
#print(tomin)
#print(type(tomin))
tohr = time.strftime("%H", time.localtime())
#print(tohr)
#print(type(tohr))
#today = time.strftime("%d", time.localtime())
if int(tohr) == 0:
    tarhr = [str(22), str(23), str(0)]
    #tarday = [today, str(int(today)-1)]
elif int(tohr) == 1:
    tarhr = [str(23), str(0), str(1)]
    #tarday = [today, str(int(today)-1)]
else:
    tarhr = [tohr, str(int(tohr)-1),str(int(tohr)-2)]
    #print(tarhr)
    #print(type(tarhr))
    #tarday = today

for d in soup.find(class_="part_list_2").find_all('h3'):
    [dhr, dmin] = d.find(class_="date").text.split(' ')[1].split(':')
    #print([dhr, dmin])
    if int(tohr) == 0:
        if dhr in tarhr:
            if dhr == '22':
                if int(dmin) > int(tomin):
                    print(d.find(class_="date").text, d.find_all('a')[-1].text)
                else:
                    break
            else:
                print(d.find(class_="date").text, d.find_all('a')[-1].text)
        else:
            break
    elif int(tohr) == 1:
        if dhr in tarhr:
            if dhr == '23':
                if int(dmin) > int(tomin):
                    print(d.find(class_="date").text, d.find_all('a')[-1].text)
                else:
                    break
            else:
                print(d.find(class_="date").text, d.find_all('a')[-1].text)
        else:
            break
    else:
        #print('-----------')
        if dhr in tarhr:
            #print('1')
            if dhr == str(int(tohr)-2):
                if int(dmin) > int(tomin):
                    print(d.find(class_="date").text, d.find_all('a')[-1].text)
                else:
                    break
            else:
                print(d.find(class_="date").text, d.find_all('a')[-1].text)
        else:
            break
        
        
    

# 關閉瀏覽器
browser.quit();
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
# 打開瀏覽器
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import Select
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

import time
today = time.strftime("%d", time.localtime()) #抓今日
selectSite = Select(browser.find_element_by_id("selD"))
selectSite.select_by_value(str(int(today)-3))
browser.find_element_by_id('button').click()

# 每個兩秒鐘自動往下滑
for i in range(30):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")

# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
#print(soup)
for d in soup.find(class_="part_list_2").find_all('h3'):
    #print(d)
    dday = d.find(class_="date").text.split(' ')[1].split(':')[0]
    #print(dday)
    #print(d.find(class_="date").text, d.find_all('a')[-1].text)
    if dday in ['15','16','17']:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    elif int(dday) < 15:
        break

# 關閉瀏覽器
browser.quit();
```
