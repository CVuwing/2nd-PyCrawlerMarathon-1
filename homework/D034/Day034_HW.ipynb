## 作業目標

* 目前程式中的 proxy_ips 是手動輸入的，請根據 https://free-proxy-list.net/ 寫一個可自動化抓取可用 Proxy 的 proxy_ips。




```python
from bs4 import BeautifulSoup
import requests
import random
```

```python
resp = requests.get('https://free-proxy-list.net/')
soup = BeautifulSoup(resp.text, 'html5lib')
aa = soup.find(class_='container')
bb = soup.find(class_='table table-striped table-bordered').find_all('tr')
```

```python
proxy_ips = []
bb.pop()
for cc in bb[1:]:
    #print(cc)
    ipS = cc.td.get_text()+':'+cc.find_all('td')[1].get_text()
    proxy_ips.append(ipS)
    #print(type(cc.td.string))
    #print(cc.find(class_='hx').get_text())
    http = cc.find(class_='hx').get_text()
    if http == 'yes':
        proxy_ips.append(ipS)
    #print(ipS)
#print(bb[1:2])
```

```python
headers = {'Referer': 'http://ip.filefab.com/index.php',
           'Sec-Fetch-Dest': 'empty', 
           'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'}
for i in range(10):
    ip = random.choice(proxy_ips)
    print('Use', ip)
    try:
        resp = requests.get('http://ip.filefab.com/index.php', headers = headers,
                        proxies={'http': ip, 'https': ip}, timeout=10)
        soup = BeautifulSoup(resp.text, 'html5lib')
        print(soup.find('h1', id='ipd').text.strip())
    except:
        print('Fail')
```
