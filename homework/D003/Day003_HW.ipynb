### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
import xmltodict

with open('./data/64_72hr_CH.xml',encoding='utf-8') as fh:
    f=fh.read()
    #print(f)
    
d = dict(xmltodict.parse(f))
print(type(d))
#print(d)
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

datasetDescription = d['cwbopendata']['dataset']['locations']['location']#[0]['locationName']
#print(datasetDescription)

areas=[]
for area in datasetDescription:
    #print(area['locationName'])
    areas.append(area['locationName'])
print('共',str(len(areas)),'區有溫度資料\n有：',areas)
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

#datasetDescription = d['cwbopendata']['dataset']['locations']['location']
#print(datasetDescription)

aal_w={}
for area in datasetDescription:
    #print(area['locationName'])
    aal_w[area['locationName']]=[area['weatherElement'][0]['time'][0]['dataTime'],
                                 area['weatherElement'][0]['time'][0]['elementValue']['value']+
                                 area['weatherElement'][0]['time'][0]['elementValue']['measures']]
print(aal_w)
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

#datasetDescription = d['cwbopendata']['dataset']['locations']['location']
a1=datasetDescription[0]['weatherElement'][0]
#print(a1)

a1_w={}
for aw in a1['time']:
    #print(aw['dataTime'], aw['elementValue']['value'], aw['elementValue']['measures'])
    a1_w[aw['dataTime']]=[aw['elementValue']['value']+ aw['elementValue']['measures']]
print(a1_w)
```
